```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your data is already loaded into the 'df' DataFrame

# 1. Group by source_city and departure_time and count occurrences
heatmap_data = df.groupby(['source_city', 'departure_time']).size().reset_index(name='count')

# 2. Pivot the data
pivot_table = heatmap_data.pivot_table(index='source_city', columns='departure_time', values='count').fillna(0)

# 3. Create the heatmap
plt.figure(figsize=(12, 8)) # Adjust figure size as needed
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlGnBu') # annot=True to show counts, fmt='d' for integer format
plt.title('Heatmap of Flights by Source City and Departure Time')
plt.xlabel('Departure Time')
plt.ylabel('Source City')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Get value counts for 'airline'
airline_counts = df['airline'].value_counts()

## Create a pie chart for 'airline'
plt.figure(figsize=(4, 4)) # Set figure size for better readability
plt.pie(airline_counts, labels=airline_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Flights by Airline')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

## Get value counts for 'source_city'
source_city_counts = df['source_city'].value_counts()

## Create a pie chart for 'source_city'
plt.figure(figsize=(4, 4)) # Set figure size
plt.pie(source_city_counts, labels=source_city_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('tab10'))
plt.title('Distribution of Flights by Source City')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Method 1: Using Seaborn Countplot with hue
plt.figure(figsize=(14, 8)) # Adjust figure size for better readability
sns.countplot(data=df, x='source_city', hue='airline', palette='viridis')
plt.title('Number of Flights by Source City and Airline')
plt.xlabel('Source City')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Airline')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group by source_city and destination_city and calculate the mean of the 'price'
average_price_by_route = df.groupby(['source_city', 'destination_city'])['price'].mean().reset_index()

## Pivot the data to create a matrix suitable for a heatmap
# Index will be source_city, columns will be destination_city, values will be the average price
heatmap_data_price = average_price_by_route.pivot_table(index='source_city',
                                                        columns='destination_city',
                                                        values='price').fillna(0) # Fill missing routes with 0 or NaN

## Create the heatmap
plt.figure(figsize=(12, 8)) # Adjust figure size as needed
sns.heatmap(heatmap_data_price, annot=True, fmt=".0f", cmap="YlGnBu") # annot=True to show average prices
plt.title('Average Flight Price by Source City and Destination City')
plt.xlabel('Destination City')
plt.ylabel('Source City')
plt.show()
```
