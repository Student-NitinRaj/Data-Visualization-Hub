```python
# Calculate total uninjured per country
uninjured_by_country = data.groupby('Country')['Total.Uninjured'].sum().sort_values(ascending=False)

# Select the top N countries for better visualization
top_n = 20  # You can adjust this number
top_countries_uninjured = uninjured_by_country.head(top_n)

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries_uninjured.index, y=top_countries_uninjured.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total Uninjured Individuals')
plt.title(f'Top {top_n} Countries by Total Uninjured Individuals')
plt.tight_layout()
plt.show()
```
